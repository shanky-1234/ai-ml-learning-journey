```python
#Encoding Part

X_One_Encode = pd.get_dummies(X,columns=['cut','color','clarity'],drop_first=True)
X_One_Encode
```

```python
bool_cols = X_One_Encode.select_dtypes(include='bool').columns
X_One_Encode[bool_cols] = X_One_Encode[bool_cols].astype(int)
```

```python
X_One_Encode
```

```python
column = ['carat','depth','table','x','y','z','volume','face_area']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_One_Encode[column] = scaler.fit_transform(X_One_Encode[column])
X_One_Encode
```

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
columns=['cut','color','clarity']

X_label = X

for i in columns:
    X_label[i] = encode.fit_transform(X_label[i])
```
